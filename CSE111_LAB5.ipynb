```python
#Task_1
import math
class Circle:
    def __init__(self,radius):
        self.__radius=radius
    def area(self):
        pi=math.pi
        return pi*self.__radius*self.__radius
    def setRadius(self,radius):
        self.__radius=radius
    def getRadius(self):
        return self.__radius       
c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
```

```python
#Task_2
class Triangle:
    def __init__(self,base,height):
        self.__base=base
        self.__height=height
    def setBase(self,base):
        self.__base=base
    def getBase(self):
        return self.__base
    def setHeight(self,height):
        self.__height=height
    def getHeight(self):
        return self.__height
    def area(self):
        return 1/2*self.__base*self.__height
    
t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#Task_6
class Passenger:
  def __init__(self,name,start="",end=""):
    self.__name=name
    self.__start=start
    self.__end=end
    self.__fair=0
  def setName(self,a):
    self.__name=a
  def getName(self):
    return self.__name
  def setFrom(self,a):
    self.__start=a
  def getFrom(self):
    return self.__start
  def setTo(self,a):
    self.__end=a
  def getTo(self):
    return self.__end 
  def setFair(self,a):
    self.__fair=a
  def getFair(self):
    return self.__fair
class Train:
  def __init__(self,a,*b):
    self.__name=a
    self.__road=b
    self.__passenger=[]
    print(f"Welcome aboard on {a}")
    print(f"Start: {b[0]}")
    print(f"Destination: {b[-1]}")
  def addPassenger(self,*r):
    for i in r:
      print(f"{i.getName()} welcome aboard")
      self.__passenger.append(i)
      if i.getFrom()=="":
        i.setFrom(self.__road[0])
        i.setTo(self.__road[-1])
      elif i.getFrom()=="":
        i.setFrom(self.__road[-1])
      fair=0
      for j in range(len(self.__road)):
        if self.__road[j]==i.getFrom():
          for x in range(j+1,len(self.__road)):
            fair+=100
            if self.__road[x]==i.getTo():
              break
          break
      i.setFair(fair)
  def allPassengerDetails(self):
    for i in self.__passenger:
      print(f"Name: {i.getName()},Start: {i.getFrom()},Destination: {i.getTo()},Fair: ${i.getFair()}")
     
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print('1========================')
p1 =Passenger('Naruto')
t1.addPassenger(p1)
p2 = Passenger('Sasuke','Manhattan')
p3 = Passenger('Hinata','Manhattan','Brooklyn')
print('2========================')
t1.addPassenger(p2,p3)
print('3========================')
t1.allPassengerDetails()
print('4========================')
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print('5========================')
p4 =Passenger('Max','London','Brussels')
p5 = Passenger('Eleven','Paris')
p6 = Passenger('Mike','Brussels')
t2.addPassenger(p4,p5,p6)
print('6========================')
t2.allPassengerDetails()
```

```python
#task7
class BracuStudent:
    def __init__(self, name, home):
        self.name = name
        self.home = home
        self.bus_pass = False

    def get_pass(self):
        self.bus_pass = True
        print(f"{self.name} got the bus pass.")

    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.bus_pass}")


class BracuBus:
    def __init__(self, route, max_capacity=2):
        self.route = route
        self.max_capacity = max_capacity
        self.passengers = []

    def board(self, *students):
        if len(students) == 0:
            print("No passenger!")
        else:
            for student in students:
                if not student.bus_pass or student.home != self.route:
                    print("You don't have bus pass!\nYou got on wrong bus!")
                elif len(self.passengers) >= self.max_capacity:
                    print("Bus is full!")
                    break
                else:
                    self.passengers.append(student.name)
                    print(f"{student.name} boarded the bus.")

    def show_details(self):
        print(f"Bus Route: {self.route}")
        print(f"Passengers Count: {len(self.passengers)} (Max: {self.max_capacity})")
        print(f"Passengers On Board: {self.passengers}")
st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#task8
class Library:
    def __init__(self, location, book_availability):
        self.location = location
        self.borrower_info = {}
        self.book_availability = book_availability
        
    def details(self):
        print(f"{self.location} Library details")
        print("Borrower details:")
        print(self.borrower_info)
        print("Books availability:")
        print(self.book_availability)
        
class Reader:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = {}
        
    def borrow(self, library, *book_types):
        for book_type in book_types:
            if book_type not in library.book_availability:
                print(f"{book_type} books are not available at the moment.")
                continue
            if library.book_availability[book_type] == 0:
                print(f"No {book_type} book is available at the moment.")
                continue
            if self.name not in library.borrower_info:
                library.borrower_info[self.name] = 0
            if library.borrower_info[self.name] + 1 > 5:
                print("You cannot borrow more than 5 books.")
                continue
            library.book_availability[book_type] -= 1
            library.borrower_info[self.name] += 1
            if book_type not in self.borrowed_books:
                self.borrowed_books[book_type] = 0
            self.borrowed_books[book_type] += 1
            print(f"{book_type} book is borrowed successfully.")
        
    def readerInfo(self, book_type=None):
        if book_type is None:
            print(f"{self.name}, you have {sum(self.borrowed_books.values())} book(s) with you.")
            for book_type, num_books in self.borrowed_books.items():
                print(f"Books on {book_type}: {num_books}")
        else:
            num_books = self.borrowed_books.get(book_type, 0)
            print(f"{self.name}, you have {num_books} {book_type} book(s) with you.")
L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
